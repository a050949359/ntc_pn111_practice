```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://movies.yahoo.com.tw/"
```

```python
html = requests.get(url)
```

```python
soup = bs(html.text,"html.parser")
movies = soup.find('ul',{'class':'movie_ind_list'}).find_all('li')
for movie in movies:
    a_tags = movie.find('a')
    print(movie.find('h3').text,a_tags['data-ga'].split(',')[2][1:-2])
```
