```python
import pandas as pd

#可以利用類似的手法爬取線上可下載資源的資料
base_url = "https://data.taipei/api/getDatasetInfo/downloadResource?id={}&rid={}"
_id = "2f238b4f-1b27-4085-93e9-d684ef0e2735"
rid = "ea731a84-e4a1-4523-b981-b733beddbc1f"
csv_url = base_url.format(_id, rid)
df = pd.read_csv(csv_url, encoding='big5')
#看前5筆資料
df.head()
```

```python
df.columns = ['發生年', '發生月', '發生日', '發生時', '發生分', '處理別', '區序', '肇事地點', '死亡人數', '受傷人數',
       '當事人序', '車種', '性別', '年齡', '受傷程度', '天候', '速限', '道路型態', '事故位置']

year = 107
month = 6
old = 30
location = "07信義區"
speed = 60
df_q = df.query("發生年 == @year & 發生月 == @month & 年齡 < @old & 區序 == @location & 速限 < @speed")
happend = df_q[['發生年','發生月']]
other = df_q[['肇事地點','速限','死亡人數','受傷人數','性別','年齡']]
happend.join(other)
```
