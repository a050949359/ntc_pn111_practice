```python
import requests
from bs4 import BeautifulSoup as bs

cwb_url = "https://e-service.cwb.gov.tw/HistoryDataQuery/DayDataController.do?command=viewMain&station=466920&stname=%25E8%2587%25BA%25E5%258C%2597&datepicker=2020-08-24"
```

```python
cwb = requests.get(cwb_url)
cwb.encoding = 'utf-8'
```

```python
soup = bs(cwb.text, "html.parser")
all_tag = soup.find('tbody').find_all('tr')[3:]

for i in range(len(all_tag)):
    print([text for text in all_tag[i].stripped_strings])
```
