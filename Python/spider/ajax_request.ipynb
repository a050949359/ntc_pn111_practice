```python
import requests,json
from bs4 import BeautifulSoup as bs

url = "https://www.kkday.com/zh-tw/product/ajax_productlist/?country=&city=A01-001-00017&keyword=%E5%8F%B0%E7%81%A3%E5%85%A9%E5%A4%A9%E4%B8%80%E5%A4%9C&availstartdate=&availenddate=&cat=&time=&glang=&sort=rdesc&page=1&row=10&fprice=*&eprice=*&precurrency=TWD&csrf_token_name=3e5a04e25c7f9700a14d72abc753899d"
```

```python
ajax = requests.get(url)
```

```python
ajax_content = ajax.json()
ajax_content['data']

for data in ajax_content['data']:
    print(str(data['price']) +" "+ data['name'])
```
