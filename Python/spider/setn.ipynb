```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://www.setn.com/ViewAll.aspx?PageGroupID=0"

html = requests.get(url)
```

```python
soup = bs(html.text,"html.parser")

for news in soup.find('div',{'class':'row NewsList'}):
    if(news.find('h3') != -1):
        print(news.find('h3').find('a').text)
        print('https://www.setn.com'+news.find('h3').find('a')['href'])
```
