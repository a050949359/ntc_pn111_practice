```python
iris = pd.read_csv('iris.csv')
iris
```

```python
filter1 = iris['target'] == 'setosa'
iris[filter1]
```

```python
filter2 = iris['sepal_length'] > 6
filter3 = iris['sepal_width'] >3.5

iris[filter2 & filter3]
```

```python
filter4 = iris['sepal_length'].between(4,4.8)
iris[filter4]

filter_setosa = iris['petal_width'] < 1
iris_setora = iris[filter_setosa]
iris_setora.to_csv(r'iris_setosa.csv')

filter_versicolor = iris['petal_width'].between(1,1.7)
iris_versicolor = iris[filter_versicolor]
iris_versicolor.to_csv(r'iris_versicolor.csv')

filter_virginica = iris['petal_width'] > 1.7
iris_virginica = iris[filter_virginica]
iris_virginica.to_csv(r'iris_virginica.csv')

```
