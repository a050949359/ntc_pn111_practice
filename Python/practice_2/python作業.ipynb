```python
import pandas as pd

skillCounter = {}

data = pd.read_csv('104行銷企劃.csv')
# print(data.info())
print(data.info.__doc__)
```

```python
skills = data['skills']
skillArr = skills[0].split("、")

for i in range(len(skills)):
    skillArr = skills[i].split("、")
    for skill in skillArr:
        if skill == "不拘":
            continue
        elif skill in skillCounter:
            skillCounter[skill] += 1
        else:
            skillCounter[skill] = 1

skillCount = pd.DataFrame(list(skillCounter.items()), columns=['skillName', 'count'])
skillCount.sort_values('count', ascending=False).head(10)
```
